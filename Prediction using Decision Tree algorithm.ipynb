```python
#Importing the Libraries:

import pandas as pd
import seaborn as sns
import sklearn.datasets as datasets
```

```python
#Loading the iris dataset:

iris = datasets.load_iris()
```

```python
#forming the dataframe:

df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

```python
sns.pairplot(df)
```

```python
sns.barplot(x='petal length (cm)',y='petal width (cm)',data=df, palette="rainbow_r")
```

```python
sns.barplot(x='sepal length (cm)',y='sepal width (cm)',data=df, palette="rainbow_r")
```

```python
# Defining the decision tree algorithm

from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)

print('Decision Tree Classifer Created Sucessfully!')
```

```python
# Import necessary libraries for graph viz

from six import StringIO 
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph

dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names, filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
